```python
import matplotlib.pyplot as plt
import waf_torch.window as atomic_window
import numpy as np
```

```python
plt.rcParams.update({'font.size': 16})
cmap = plt.get_cmap('viridis')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
```

```python
atomic_window.window_list
```

```python
fig = plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.plot(atomic_window.up(100),label=r'$\mathrm{up}$',color=colors[0])
plt.plot(atomic_window.upm(100,m=3),label=r'$\mathrm{up}_3$',color=colors[1])
plt.plot(atomic_window.ha(100,a=3),label=r'$\mathrm{h}_3$',color=colors[2])
plt.plot(atomic_window.xin(100,n=3),label=r'$\mathrm{\Xi}_3$',color=colors[3])
plt.plot(atomic_window.fupn(100,n=3),label=r'$\mathrm{fup}_3$',color=colors[4])
plt.plot(atomic_window.chan(100,a=2,n=3),label=r'$\mathrm{ch}_{2,3}$',color=colors[5])
plt.plot(atomic_window.fipan(100,a=2,n=1),label=r'$\mathrm{fip}_{2,1}$',color=colors[6])
plt.plot(atomic_window.fpmn(100,m=2,n=1),label=r'$\mathrm{fp}_{2,1}$',color=colors[7])
plt.title(r'$\mathrm{mode}=max$')
plt.legend()

plt.subplot(1,2,2)
plt.plot(atomic_window.up(100,mode='area'),label=r'$\mathrm{up}$',color=colors[0])
plt.plot(atomic_window.upm(100,m=3,mode='area'),label=r'$\mathrm{up}_3$',color=colors[1])
plt.plot(atomic_window.ha(100,a=3,mode='area'),label=r'$\mathrm{h}_3$',color=colors[2])
plt.plot(atomic_window.xin(100,n=3,mode='area'),label=r'$\mathrm{\Xi}_3$',color=colors[3])
plt.plot(atomic_window.fupn(100,n=3,mode='area'),label=r'$\mathrm{fup}_3$',color=colors[4])
plt.plot(atomic_window.chan(100,a=2,n=3,mode='area'),label=r'$\mathrm{ch}_{2,3}$',color=colors[5])
plt.plot(atomic_window.fipan(100,a=2,n=1,mode='area'),label=r'$\mathrm{fip}_{2,1}$',color=colors[6])
plt.plot(atomic_window.fpmn(100,m=2,n=1,mode='area'),label=r'$\mathrm{fp}_{2,1}$',color=colors[7])
plt.title(r'$\mathrm{mode}=area$')
plt.legend()

plt.show()
```
